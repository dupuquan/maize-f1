```python
import pandas as pd
import itertools
import numpy as np
```

```python
df=pd.read_excel('parents_clean (9).xlsx',keep_default_na=False)
```

```python
name_id= df[['id']].copy()
ColNames = df.columns
EMF0901135939 = df[['EMF0901135939']].copy()#0
EMF2000238400 = df[['EMF2000238400']].copy()#1
EMF0700126067 = df[['EMF0700126067']].copy()#2
EMF2000893574 = df[['EMF2000893574']].copy()#3
EMF2000069773 = df[['EMF2000069773']].copy()#4
EMF0500586084 = df[['EMF0500586084']].copy()#5
EMF2000892076 = df[['EMF2000892076']].copy()#6
EMF2000178475 = df[['EMF2000178475']].copy()#7
EMF0700676717 = df[['EMF0700676717']].copy()#8
EMF2001235465 = df[['EMF2001235465']].copy()#9
```

```python
dfs = [EMF0901135939, EMF2000238400, EMF0700126067, EMF2000893574, EMF2000069773, EMF0500586084, EMF2000892076, EMF2000178475, EMF0700676717, EMF2001235465]
```

```python
combined_dfs = []
for combo in itertools.combinations(dfs, 2):
    combined_dfs.append(pd.concat(combo, axis=1))
```

```python
result_df = pd.concat(combined_dfs, axis=1)
# 获取列名
column_names = result_df.columns.tolist()

# 生成所有不重复的列名组合
column_combinations = list(itertools.combinations(column_names, 2))
column_combinations = sorted(set(column_combinations))
```

```python
for i in column_combinations:
    for j in range(1,2):
        if (i[j]==i[j-1]):
            column_combinations.remove(i)
        
```

```python
cleaned_list = list(set([tuple(sorted(item)) for item in column_combinations]))
```

```python
print(cleaned_list)
```

```python
new_data=[]
for i in range(0,len(cleaned_list)):
    for column in range(1,10):
        if cleaned_list[i][1]==df.columns[column]   or cleaned_list[i][0]==df.columns[column]:
            new_data.append(df.iloc[:,column])
```

```python
print(new_data)
```

```python
# 设置每块写入的行数
chunk_size = 20  # 可以根据实际情况调整这个值

# 打开文件用于写入
with open('F1.txt', 'w') as file:
    # 计算需要分成的块数
    num_chunks = len(new_data) // chunk_size + (1 if len(new_data) % chunk_size else 0)
    
    # 分块写入数据
    for i in range(num_chunks):
        start_index = i * chunk_size
        end_index = (i + 1) * chunk_size
        chunk_data = new_data[start_index:end_index]
        
        # 将当前块的数据转换为字符串并写入文件
        file.write('\n'.join(chunk_data) + '\n')

print("所有数据已成功导出到F1.txt文件中。")
```
