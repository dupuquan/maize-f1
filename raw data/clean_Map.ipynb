需要使用动态规划算法，删除遗传紊乱的位点

```python
def longest_increasing_subsequence(data):
    n = len(data)
    dp = [1] * n
    for i in range(1, n):
        for j in range(i):
            if data[i] > data[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    max_length = max(dp)
    max_index = dp.index(max_length)
    lis = [data[max_index]]
    for i in range(max_index - 1, -1, -1):
        if data[i] < data[max_index] and dp[i] == dp[max_index] - 1:
            lis.append(data[i])
            max_index = i
    lis.reverse()
    return lis
```
